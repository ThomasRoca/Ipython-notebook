## <span style="color:#00acee">1. Line chart generator using Bokeh

```python
# import library
from IPython.display import HTML
import pandas as pd
import json
import numpy as np
from bokeh.plotting import figure, show, output_file, vplot
from bokeh.models import HoverTool, BoxSelectTool
import os
import sys
#!pip install bokeh --upgrade

'''
##### Delete last version of it #######
try:
    os.remove("lineGraph.html")
except OSError:
    pass
#######################################
'''

#######################################
#print ("Please give me a country list - iso3 comma separated:")
IndicatorID = input("Please give me an indicator:")
Country_list = input("Enter country list - space seperated:").split(" ")
Year_list=input("Please give me a year list - space separated:")
#######################################


################ Transform the input ###################

#convert input list from str to int
Year_list = [int(x) for x in  Year_list.split()]

#########################################################       
        
#Set year range for graph (i.e the x axis range (-1;+1))
#Save last item position in Year_list
ly=len(Year_list)-1
#save Range
Rmin=Year_list[0]-1
Rmax=Year_list[ly]+2

#Format to fit in the url
# 1. country list 
c=Country_list[0]
lc=len(Country_list)
for item in range(1, lc):
      c=c+","+Country_list[item]

# 2. Year list         
y=str(Year_list[0])    
ly=len(Year_list)
for itemY in range(ly+1-ly, ly):
     y=y+","+str(Year_list[itemY])

# 3. Indicator
indicator_id=str(IndicatorID)    
#print("The following data will be presented: Indicator id: "+indicator_id+ " | country:"+c+" | year:"+y)

# 4. Generate random colors
import random
for i in range(0,lc):
    vars()['color_'+str(i)]= "#%06x" % random.randint(0,0xFFFFFF)

#  4.1 or use formatted ones
color_1="#00adef" #blue1
color_2="#ffdf51" #Yellow
color_3="#a6ca6d" #Green
color_4="#ad46d6" #purple
color_5="#f26a2e" #Orange
color_6="#6e9fc5" #blue2
color_7="#f4bb90" #pale pink

################ set URL #####################
url="http://ec2-52-1-168-42.compute-1.amazonaws.com/version/1/country_code/"+c+"/indicator_id/"+indicator_id+"/year/"+y+"?structure=icy"
print('Link to API:',url)
################ get the data #####################
#load dataframe:
try:
    df=pd.read_json(url)
except IOError:
    sys.exit("ATTENTION: We are sorry but the data your requested for this indicator/year is not available")


lc=len(Country_list)

#Save country names
lc=len(Country_list)
for b in range(0, lc):
     cname=Country_list[b]
     vars()['cname'+str(b)]=df['country_name'][cname]            

#Save the variable name as title 
title=df['indicator_name'][indicator_id]

######################################### Save the data in lists ##############################################

# Record the first value for each observation [NB. to deal wit]
for c in range(0,lc):  
       vars()['data'+str(c)]=df['indicator_value'][indicator_id][Country_list[c]][str(Year_list[0])]

# Add the other value for each observation
for c in range(0,lc):  
    for i in range(1,ly):
        vars()['data'+str(c)]=vars()['data'+str(c)]+' '+df['indicator_value'][indicator_id][Country_list[c]][str(Year_list[i])]

#store and convert variable string list in float list
for c in range(0,lc):  
      vars()[('data'+str(c))]=[float(i) for i in vars()[('data'+str(c))].split()]

        
######################################### Set Graph options ##############################################

#TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"
TOOLS=[BoxSelectTool(), HoverTool()]
graph = figure(title=title, tools=TOOLS, x_range=(Rmin,Rmax), plot_width=750, plot_height=600)

#set different options:
graph.xaxis.major_label_orientation = 3.14 / 3.0
graph.axis.minor_tick_in = 0    #Set size of inner tick
graph.axis.minor_tick_out = 0   #Set size of inner tick

#Set tooltip
hover = graph.select(dict(type=HoverTool))
hover.tooltips = [
    ("x", "@x"),
    ("y", "@y"),
]
#Set legend placement
#graph.legend.orientation = "top_right"  
graph.legend.legend_spacing = 3
graph.legend.legend_padding = 5

######################################### Input data to the Graph ##############################################
for c in range(0,lc):
    graph.line(Year_list, vars()[('data'+str(c))], legend= vars()['cname'+str(c)], color=vars()[('color_'+str(c))],line_width=2)
    graph.circle(Year_list, vars()[('data'+str(c))], legend= vars()['cname'+str(c)], color=vars()[('color_'+str(c))],size=6)  

    
show(vplot(graph))  # open a browser
output_file("lineGraph.html")

#embed results as html
HTML('''<iframe src="lineGraph.html" scrolling="no"  frameborder="0" width="80%" height="655px"></iframe>''')

```
