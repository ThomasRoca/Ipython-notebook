Although, crime records have long been archived, the use of crime and offense data has blossomed with the progress of data science. Statisticians can now help police forces predicting violence momentum and adapting police unit deployment to the probability of violence outbursts during certain hours of the day or across certain neighborhoods.
Although, one may imagine criminal events pop-up randomly, crime statistics analysis show that certain factors are accurate predictors of violence (see Alves LGA & al, 2013 [i] or McCue C, 2015 [ii], keegan BC, 2014). Meteorological conditions, for example, correlate with homicides. Cold weather is known to be associated with less crime, so as rainy days, or hot ones - basically, when people stay home... This winter (2015) for example, due to the severe cold condition, New York city recorded its lower murder rate for decades. In February, no murder was reported in New York city for <a href="http://www.theguardian.com/us-news/2015/feb/14/record-new-york-city-freeze-hits-12-days-as-experts-cite-extreme-cold" target="_blank">12 days</a>, a record since the city started analyzing crime data, 30 years ago.


As fancy – or scary - as might sound big data and predictive crime unit (e.g. using social network mining to identify crime predictors), the raw material: crime records (i.e. the dependent variable) is known to suffer from different bias.(see Skogan WG, 1974 [iv]; Smith A, 2006 [v] or Carrell S. E. & Hoekstra M, 2012[vi]).

In this blog-post we analyze crime data, provided by the police forces of Cartagena (Colombia), as part of the hackathon that took place during the <a href="http://http://www.cartagenadatafest2015.org/" target="_blank">Cartagena Data Festival</a>, organized by <a href="http://www.datapopalliance.org/" target="_blank">Data-Pop Alliance</a>, <a href="http://www.odi.org/" target="_blank">ODI</a>, <a href="http://www.undp.org/" target="_blank">UNDP</a>, and others.


*Sicariato* - murder perpetrated by the use of  hired  gunmen - is the most common type of homicide in Cartagena, whether victims are men (752 victims between 2007 an 2013) or women (28 victims). While *Riña* - fights - are the second cause of death among man population, domestic violence comes in 2nd position for women casualty (23 victims).
Looking more closely to domestic violence, we figured out that our dataset contains uncommon results. The records actually reports that domestic violence affect equally men and women. Actually, in absolute value more men (24 deaths) that women were killed in the context of domestic violence.

As a benchmark, we reproduced below a table summarizing domestic violence by gender ,published by the New York City Police department. In New York City, crime records appear more in line with intuition (i.e. women more likely to be victim of domestic violence) as 51.4% of domestic violence victims were women (vs. 25.1% for men). This represents, in absolute value, a 2:1 ratio - 437 women victims vs. 214 men.

At this stage two assumptions may be formulated:

- In Cartagena women are strong, and you'de better not mess with them !
- Our dataset may contain misreported case of domestic violence..


## *Conclusion:* ##
Although we have no doubt on how strong - while charming - women are in Cartagena, the data shows very uncommon results when it comes to domestic violence. A rapid benchmark with New York City shows that it is unlikely that men are more victim of domestic violence than women. This cast doubt on the domestic violence reporting in the dataset we studied. We suggest a possible misreporting of domestic violence towards women. 


This very rapid and partial analysis (the result of a 2-3h workload) reminds us that however hype Data science is - so as big data is- a sound analysis also relies on sound 'small' data. Resorting to innovative data such as social mining etc. to help predicting and analysing "small data" can be pointless if we don't benefit from good quality "small data" in the first place...
